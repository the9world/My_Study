선형적(linear)이란, '직선(line) 모양  
두 변수 사이에 직선 형태의 관계가 있는 것을 의미
떤 사람의 키와 몸무게 사이의 관계는 직선으로 표현할 수 있습니다.  
이 때, 키가 1cm 증가할 때 몸무게는 항상 a kg 증가한다는 것을 의미합니다.  
즉, 두 변수 간의 변화가 일정하고 예측 가능한 패턴을 가지고 있습니다.

비선형적(nonlinear)이란 직선 모양이 아닌 것을 의미  
두 변수 사이에 직선 형태가 아닌 곡선 형태의 관계가 있는 것을 의미  
어떤 사람의 나이와 소득 사이의 관계는 곡선으로 표현할 수 있습니다.  
이 때, 나이가 1살 증가할 때 소득이 b만원 증가하는 것이 아니라, 나이가 증가함에 따라 소득 증가폭이 달라지는 것을 의미합니다.  
즉, 두 변수 간의 변화가 일정하지 않고 예측하기 어려운 패턴을 가지고 있습니다.

미분(그래디언 디센트)
y축 코스트
좌우 극한이 맞지 않으면 미분이 불가

미분의 원리는 함수의 변화율을 구하는 것입니다. 함수의 변화율은 한 순간에서의 기울기를 의미합니다. 즉, 함수의 한 점에서의 변화율은 그 점에서의 기울기를 의미합니다.

미분의 기본적인 공식은 다음과 같습니다.

f'(x) = lim (h → 0) (f(x+h) - f(x)) / h

여기서 f'(x)는 함수 f(x)의 도함수를 나타내며, x를 중심으로 오른쪽 방향으로 아주 작은 h만큼 이동한 f(x+h)와 x에서의 함수값 f(x)의 차이를 h로 나눈 것입니다. 이 공식은 함수 f(x)의 한 점에서의 변화율(기울기)을 구하는 것을 의미합니다.

미분을 통해 구한 도함수 값은 x에서의 순간 기울기, 즉 접선의 기울기를 나타냅니다. 따라서 미분은 함수의 극값(최대, 최소)이나 볼록성, 오목성 등을 알아내는 데에 매우 유용하게 사용됩니다.

미분을 더 쉽게 이해하기 위해서는 기초 수학 개념인 변화율(rate of change)을 이해하는 것이 중요합니다.

변화율이란 시간, 거리 등 어떤 변수의 변화량에 따라 다른 변수가 얼마나 변화하는지를 나타내는 지표입니다. 예를 들어, 자동차가 1시간 동안 60km에서 100km로 가속할 경우, 이동한 거리의 변화량은 40km이며, 이에 따른 속도의 변화율은 40km/h 입니다.

미분은 이러한 변화율의 개념을 이용하여 함수의 순간 변화율을 계산하는 것입니다. 이를 계산하기 위해서는 함수의 미소 변화량이 얼마나 되는지를 계산해야 합니다. 이 미소 변화량은 기본적으로 함수의 접선의 기울기와 같습니다. 따라서 미분을 계산하기 위해서는 함수의 접선의 기울기를 계산하는 것이 중요합니다.

미분은 다음과 같은 수식으로 나타낼 수 있습니다.

$f'(x) = \lim_{h\to0} \frac{f(x+h)-f(x)}{h}$

여기서 h는 미소 변화량을 의미하며, h가 0에 가까워질수록 더욱 정확한 순간 변화율을 계산할 수 있습니다.
