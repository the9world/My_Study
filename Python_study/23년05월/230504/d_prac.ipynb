## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib jupyter notebook
#jupyter notebook에서 그려주세요.
# %matplotlib qt #별도의 팝업창에서 그래프 출력(colab X)
```

- 선 그래프 line plot
- plt.plot 함수는 주어진 데이터를 그래프로 그려주는 함수,  
   인자로 데이터 리스트를 전달합니다.
- 그래프를 그리기 위해서는 matplotlib 라이브러리를 불러와야 하며,  
  plt.plot 함수를 이용하여 그래프를 그린 후에는 

```python
data = [10,40,20,60,70] # 
plt.plot(data) # data 변수에 요소로 선 그래프를 그림
```

```python
x= np.arange(10)
y= x+10
plt.plot(x,y) # x 1~10 행, y x+10 열로 선 그래프 입력
plt.show() # 그래프 출력
```

- xlim,  ylim
  - x 축의 길이와 y 축의 길이를 조정

```python
plt.xlim([0,10]) # x축의 범위를 0부터 10까지로 설정
plt.xlim([0,20]) # y축의 범위를 0부터 20까지로 설정
plt.plot(x,y) # ([])를 하는 이유는 list로 설정하려고.
plt.show()
```

```python
x = np.arange(-np.pi,np.pi,0.02) # -원주율부터 원주율까지 0.02씩 증가 (총 314개)
y1= np.sin(x) # 사인 함수를 계산 np.sin(np.pi/2)는 1을 반환
y2= np.cos(x) # 코사인 함수를 계산 np.cos(0)은 1을 반환
plt.plot(x,y1,label='sin', color=(0.1,0.3,0.5)) # RGB 값
#  x 값을 가로축, y1 세로축, label 인자를 사용하여 그래프의 범례(이름)를 지정
plt.plot(x,y2,label='cos', color='#E4D00A') # 16진수 색상코드 hex color 
#  x 값을 가로축, y2 세로축, label 인자를 사용하여 그래프의 범례(이름)를 지정
plt.legend() # plt.legend()함수는 그래프의 범례를 그래프의 오른쪽 상단에 위치
plt.show() # label(범례이름), X or y_label = 해당 행열의 이름
```

```python
plt.plot([1,2,3,4],[2,3,5,10], 'bo-')
# 'b'색상(그래프), 'o'마커(점모양), '-' 선모양
plt.xlabel('X-Axis') # x 표 행 밑 이름
plt.ylabel('Y-Axis') # y 표 열 옆 이름
```

```python
x = np.arange(10) # 1~9 배열을 x축 값에 저장
y1=x # x 값 그대로
y2=x**2-x # x^2-x
fig,axs=plt.subplots(2,1) # fig, axs 두개의 변수 생성
# fig는 전체 그림(전체), axs는 서브플롯 배열(각 그래프표), subplot(2,1)은 2행 1열 생성
fig.set_facecolor('#c79fef')
# fig.set_facolor 전체 배경 색상
axs[0].plot(x,y1) # 0번 그래프 그리기
axs[1].plot(x,y2) # 1번 그래프 그리기
axs[0].set_facecolor('pink') # 0번 그래프 배경 색상
axs[1].set_facecolor('skyblue') # 1번 그래프 배경 색상
```

- 0 ~ 360도를 나타내는 것은 60분법으로 계산을 하는 것이며, degree라고 부름
- 국제적으로 360도, 즉 1바퀴(1회전)를 2π 라디안이라고 부릅

```python
# 선 그래프 line plot
x= np.arange(-5,5,0.5) #-5부터 5까지 0.5씩 배열생성
y1= x # y1는 x와 같음
y2= x**2 # y2는 x^2
y3= np.sin(x) # x에 대한 사인 값 계산
y4= np.cos(x) # x에 대한 코사인 값 계산

plt.plot(x, y1) # 파란 선그래프
plt.plot(x, y2, marker='D') # 주황 점선그래프
plt.plot(x, y3, color= 'r') # 빨간 sin 선그래프
plt.plot(x, y4, linestyle='dashed') # 녹색 cos 점그래프
plt.show()
```

```python
# 선"표" (bar plot)
data ={'apple':21, 'banana':15, 'pear':5, 'kiwi':20}
names =list(data.keys()) # key 값들을 names 저장
values =list(data.values()) # value 값들을 values에 저장
fig,ax = plt.subplots() # ax는 단일 axs는 다중
# axs = plt.subplots(2, 1)와 같이 subplots 함수를 사용하여, 2개의 subplot을 생성한 후, 
# axs[0].plot(x1, y1)과 axs[1].plot(x2, y2)와 같이 plot 함수를 사용하여
# 각 subplot에 그래프를 그릴 수 있음, (plot=1, plots=다중)
ax.bar(names,values)
```

```python
# hist
data= np.random.rand(10000) # 0~1 무작위 10,000개
fig,ax= plt.subplots()
ax.hist(data,bins=300, facecolor='r') # bins 구간(bin)의 개수를 설정
plt.show()
```

```python
# 2D Scatter plot
n= 50
x= np.random.rand(n) # 0~1, n개의 랜덤 값
y= np.random.rand(n)
plt.scatter(x,y)
plt.show()
```
