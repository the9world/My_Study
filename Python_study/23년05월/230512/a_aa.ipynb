- 정확도(Accuracy) TP+TN / TP+TN+FP+FN  
    정확도로 전체 예측한 것 중에 올바른 예측을 얼마나 했는지  
    Error Rate = 1-Accuracy, 편향의 함정이 있음.
- 정밀도(Precision) TP / TP+FP  
    정밀도로 Positive로 예측한 것 중 실제로 맞춘 비율  
    Positive로 예측한 것 중 실제로 Positive가 얼마나 되는지  
- 민감도(Sensitivity, recall) TP / TP+FN  
    True Positive Rate로 실제 POtsitive를 얼마나 잘 예측했는지  
    Positive 중 Positive로 예측한 것이 얼마나 되는지 보여주는 지표  
- 특이도(Specificity) TN / FP+TN  
    Specificity는 Negative를 얼마나 잘 예측했는지  
    특이도를 통해서 False Positive Rate를 구할 수 있음.  
- FP rate(False Positive Rate) = 1 - Splcificity(특이도)
- F1 = 2 x (정밀도x민감도) / (정밀도+민감도)
