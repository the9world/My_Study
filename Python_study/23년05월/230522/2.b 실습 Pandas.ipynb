### first_description 이라는 변수에는, 디스크립션 컬럼의 첫번째 데이터를 저장한다.


```python
# 데이터 프레임에서 데이터를 가져오는 방법

# 1. 컬럼데이터 가져오는 방법
#    변수에 []대괄호 사용 : reviews['country']

# 2. 행과 열의 조합으로 데이터를 가져오는 방법 1
#    사람용으로 사용된 인데긋와 컬럼명으로 데이터 가져오는 방법  
#    변수에 .loc[ , ]를 사용. reviews.loc[ 0 , "country" ]

# 3. 행과 열의 조합으로 데이터를 가져오는 방법 2
#    컴퓨터용으로 사용된 인덱스로 데이터 가져오는 방법 : 숫자만 사용
#    변수에 .iloc[ , ] 를 사용, reviews.loc[ 0 , 0 ]
```

```python
first_description	= reviews["description"][0] # 데이터의 인덱스 0번째
first_description	
```

```python
first_description	= reviews.loc[0,"description"] # 데이터의 인덱스0, col명
first_description
```

```python
first_description	= reviews.iloc[0,1] # iloc는 컴퓨터가 지정한 인덱스?
first_description
```

### Italy 에서 만들어진 와인에 대해서 italian_wines 이라는 이름으로 데이터프레임을 만드시오. 


```python
# reviews["country"]=="Italy"  # boolean(bool)형태로 반환(True, False) : reviews["country"]=="Italy"을 Python에 질문 
italian_wines= reviews[reviews["country"]=="Italy"]
italian_wines
```

### 리뷰점수가 95점 이상이고, Australia와 New Zealand 에서 만들어진 와인에 대한 데이터프레임을 top_oceania_wines 이라는 이름의 변수로 저장.


```python
# reviews["points"] >= 95
# reviews['country'] == "Australia"
# reviews['country'] == "New Zealand"
# 이렇게 순서대로 풀어서 코드를 짜고 합치면 쉬움
top_oceania_wines = reviews[(reviews["points"]>=95) & ((reviews["country"] == "Australia") | (reviews["country"] == "New Zealand"))]
# top_oceania_wines = reviews.loc[(reviews["points"]>=95) & ((reviews["country"] == "Australia") | (reviews["country"] == "New Zealand"))]
top_oceania_wines
```

```python
# reviews(변수) is(에) in들어있어?
# reviews['country'].isin(["Austraila", "New Zealand"])
top_oceania_wines= reviews.loc[(reviews["points"]>=95) & (reviews['country'].isin(["Australia", "New Zealand"]))]
top_oceania_wines
```
